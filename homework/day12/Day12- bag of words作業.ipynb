## 載入套件

```python
import pandas as pd
import nltk
#nltk.download()
import numpy as np
```

## 讀取檔案

```python
dataset = pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)
corpus = dataset['Review'].values
```

### print corpus 

```python
corpus[:10]
```

### 從文本中取出所有單字

```python
whole_words = []
for sentence in corpus:
    tokenized_sentence = nltk.tokenize.word_tokenize(sentence)
    for word in tokenized_sentence:
        whole_words.append(word)
```

```python
# print out whole_words
whole_words[:10]
```

### 移除重複單字

```python
print('移除重複單字之前共有{}個單字:'.format(len(whole_words)))
```

```python
whole_words = set(whole_words)
print('共有{}個單字'.format(len(whole_words)))
```

### 建立字典使每一個單字有對應數值

```python
word_index = {}
index_word = {}
n = 0
for word in whole_words:
    word_index[word] = list(whole_words).index(word)
    index_word[n] = word
    n += 1
    
```

## 轉換句子為bag of words型式

```python
def _get_bag_of_words_vector(sentence, word_index_dic, whole_words):
    sentence = sentence
    '''創建一個vector'''
    vector = [0]*len(word_index_dic)
    
    for word in nltk.word_tokenize(sentence):
        if word in whole_words:
            print('word', word, '\n index', word_index_dic[word])
            
            '''自行填入'''
            vector[word_index_dic[word]] = 1
            
    return vector
```

### 測試 BOW
