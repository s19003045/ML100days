```python
print(sentence)
```

```python
# processing data

sentence = sentence.replace("\n", "")
sentence
```

```python
print("output 精確模式: {}".format('|'.join(jieba.cut(sentence, cut_all=False, HMM=False))))

```

## 新增字典

```python
# add words to file: if not exist, then create it 
new_words = "肖特\n裝瓶"

# 必須設定編碼方式: utf-8
with open('./new_words.txt', "w", encoding='utf8') as f:
    f.write(new_words)
# jieba.add_word(new_words)
```

```python
# verify if words writing into file

with open('./new_words.txt', "r", encoding='utf8') as file:
    data = file.read()
    print(data)
```

```python
## 讀入字典
jieba.load_userdict('./new_words.txt')
```

```python
print("output 精確模式: {}".format('|'.join(jieba.cut(sentence, cut_all=False, HMM=False))))
```

```python
# print out word by word

sentence_cut = jieba.cut(sentence, cut_all=False, HMM=False)

for i in sentence_cut:
    print(i)
```

```python
# tokenize sentence an print out word by word

sentence_token = jieba.tokenize(sentence)

for i in sentence_token:
    print(i)
```

## Taggin

```python
words = pseg.cut(sentence,)
for word, flag in words:
    print(word, flag)
```
